```python
def repl_called_x(grid, called_number):
    '''Takes a list as input. Finds in the lines of the grid all occurences of called_number.
    Replace these occurences by 0.'''

    for line in range(len(grid)):
        if called_number in grid[line]:
            for number in range(len(grid[line])):
                if grid[line][number] == called_number:
                    grid[line][number] = 'x'

    return grid
```

```python
def get_line_bingo(grid):
    'Return True if a given grid as a bingo in one of its line.'
    for line in grid:
        if line == ['x', 'x', 'x', 'x', 'x']: return True
    return False
```

```python
def get_column_bingo(grid):
    'Returns True if a given grid as a bingo in one of each column.'
    # for i in range(numbers per line)
    for i in range(len(grid[0])):
        if grid[0][i] == grid[1][i] == grid[2][i] == grid[3][i] == grid[4][i] == 'x':
            return True
    return False
```

```python
def get_winning_grid(input_grids: list[list[int]], input_numbers: list[int]) -> tuple[list, int]:
    for called_number in input_numbers:
        for grid in input_grids:
            repl_called_x(grid, called_number)

            if get_column_bingo(grid): return grid, called_number
            if get_line_bingo(grid): return grid, called_number
```

```python
def sum_winning_grid(grid: list[list[int]]) -> int:
    sum = 0
    for line in grid:
        for number in line:
            if isinstance(number, int): sum += number
    
    return sum
```

```python
def part_1(input_grids: list[list[int]], input_numbers: list[int]) -> int:
    winning_grid, last_number = get_winning_grid(input_grids, input_numbers)

    sum = sum_winning_grid(winning_grid)

    return sum * last_number
```

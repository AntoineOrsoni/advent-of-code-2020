```python
from typing import Dict, List
```

```python
with open('input') as file:
    input = file.read().splitlines()
```

```python
tree = dict()
tree['/'] = dict()
```

```python
view = tree['/']
depth = ['/']

for line in input[1:]:

    instruction = line.split(' ')

    match instruction[0]:

        ## -- Commmand -- ##
        case '$':
            match instruction[1]:
                case 'cd':
                    match instruction[2]:
                        case '..':
                            depth.pop()
                            view = tree
                            for key in depth:
                                view = view.get(key)
                        case _:
                            depth.append(instruction[2])
                            view = view[instruction[2]]
                case 'ls':
                    pass
        case 'dir':
            view[instruction[1]] = dict()

        ## -- Files -- ##
        case _:
            view[instruction[1]] = int(instruction[0])
```

```python
import pprint

pp = pprint.PrettyPrinter(indent=2)

pp.pprint(tree)
```

```python
dir_sizes = []

def get_dir_size(dir: dict) -> int:

    size = 0

    for key, value in dir.items():

        # it's a file
        if type(value) == int:
            size += value

        # it's a dir
        else:
            size += get_dir_size(value)
            dir_sizes.append({key: size})

    return size
```

```python
get_dir_size(tree)
```

```python
part_1 = 0

for dir in dir_sizes:
    for value in dir.values():
        if value < 100000:
            part_1 += value

print('answer part_1:', part_1)
```

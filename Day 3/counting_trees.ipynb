```python
with open('input', 'r') as file:
    # Saving input as a list
    input = file.read().split('\n')  
```

```python
for index, line in enumerate(input):
    print(f"{index}: {line}")
```

```python
def count_trees_squares(list, right, down):
    trees = 0
    squares = 0

    for index, line in enumerate(list):
        # print(line[(right * index) % len(line)])
        if index % down == 0:
            # print(f"value: {right * index // down} -- index: {index}")
            if line[(right * index // down) % len(line)] == '#': trees += 1
            elif line[(right * index // down) % len(line)] == '.': squares += 1
    
    return trees, squares
```

```python
trees, squares = count_trees_squares(input, 1, 1)
print(trees)

trees, squares = count_trees_squares(input, 3, 1)
print(trees)

trees, squares = count_trees_squares(input, 5, 1)
print(trees)

trees, squares = count_trees_squares(input, 7, 1)
print(trees)

trees, squares = count_trees_squares(input, 1, 2)
print(trees)
```
